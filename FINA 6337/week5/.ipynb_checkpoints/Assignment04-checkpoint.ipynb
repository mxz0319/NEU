1. Compute the percentage monthly log returns on equities and dividends expressed in percentages, defined as:
$$ re_t = 100 \times (\log P_t - \log P_{t-1}) $$
$$ rd_t = 100 \times (\log D_t - \log D_{t-1}) $$
Plot the two series. What do you notice about their time series patterns? (Note: You'll get a better looking plot by default if you use the plot function from Pandas instead of from Pyplot.)

```python
# Compute returns
return_equity = (np.log(df['P']) - np.log(df['P']).shift()).mul(100)
return_dividend = (np.log(df['D']) - np.log(df['D']).shift()).mul(100)
return_equity.plot()
plt.title('Return on Equity')
```

4. Now we'll try a couple of different models for equity returns. First, estimate an AR(6) model for equity returns. (Recall from the lecture discussion that AR models can be estimated using OLS). Comment on the coefficients.

```python
# Your code here
model = AutoReg(return_equity.dropna(),6)
ar_result = model.fit()
print(ar_result.summary())
```

The coefficients of this model are all in the good standard range.

5. Next, estimate an MA(3) model for equity returns. As we discussed in class, this cannot be done using OLS. Instead we'll use the following two step process.

(i) Estimate an AR model for $re_t$ with a constant term and $p > 3$ autoregressive terms:
$$ re_t = \phi_0 + \phi_1 re_{t-1} + \cdots + \phi_p re_{t-p} + u_t $$
The value of $p$ is up to you. Just make sure it's more than 3. After you estimate it, save the residuals. You'll need it for the next step.

```python
# Your code here
residuals = ar_result.resid
ma_model = sm.tsa.arima.ARIMA(residuals, order=(0, 0, 3))
ma_results = ma_model.fit()
ma_results.summary()
```

(ii) Next, estimate the MA(3) parameters by estimating the model:
$$ re_t = \psi_0 + \psi_1 \widehat{u}_{t-1} + \psi_2 \widehat{u}_{t-2} + \psi_3 \widehat{u}_{t-3} + v_t $$
Here, $\widehat{u}_{t-1}$ is the first lag of the residual from the previous step and so on. $v_t$ is again a random noise term. You can use OLS for this step. You'll need to be careful with the size of the dependent and independent variables above for this to work.

```python
# Your code here
residual_ma = ma_results.resid
residual_ma = residual_ma[6:]
return_equity = return_equity.dropna()
return_equity2 = return_equity[9:]

lagged_residuals = pd.DataFrame({
    'u_t-1': residual_ma.shift(1),
    'u_t-2': residual_ma.shift(2),
    'u_t-3': residual_ma.shift(3),
})
lagged_residuals = sm.add_constant(lagged_residuals)
ols_model = sm.OLS(return_equity2, lagged_residuals.dropna())
ols_results = ols_model.fit()
ols_results.summary()
```

(iii) The OLS coefficients above are the coefficients of the MA(3) model. What conclusion would you draw from these coefficients?

I think the MA(3) model has the ability to captures the impact of past periods' residuals on the current equity returns. The coefficients show the affection of the equity returns. The Positive coefficients suggest that positive residuals from the AR(6) model tend to lead to higher equity returns, while negative coefficients at t-3 could means that there is a decline in returns or as the negative number is really small, so it could be a noice term.

6. Using the same two-step procedure as above, estimate an ARMA(1,1) model of equity returns. That is, the two-step procedure is:

(i) Estimate an AR(p) model with $p > 1$ and save the residuals $\widehat{u}_t$

(ii) Use OLS with the following independent variables: 1 lag of the dependent variable; and 1 lag of the residuals from the previous step.

After that, write your conclusions.

```python
model2 = AutoReg(return_equity.dropna(),2)
ar_result2 = model2.fit()
ar_resid2 = ar_result2.resid

return_equity3 = return_equity[5:]

lagged_residuals2 = pd.DataFrame({
    'u_t-1': ar_resid2.shift(1),
    'u_t-2': ar_resid2.shift(2),
    'u_t-3': ar_resid2.shift(3),
})
lagged_residuals = sm.add_constant(lagged_residuals2)
ols_model2 = sm.OLS(return_equity3, lagged_residuals2.dropna())
ols_results2 = ols_model2.fit()
ols_results2.summary()
```

Both of the 1 lag of the dependents and residuals can tell us the impact of the past period's equity returns or residual on the current equity returns or ARMA model. A positive coefficient leads to a higher equity returns, negative means it could be a decline in returns.

7. Estimate a bivariate VAR(6) model using the equity and dividend returns.

```python
# Your code here
# Equity 
data3 = pd.concat([return_equity, return_dividend.dropna()], axis=1, keys=('Equity','Dividends'), join='inner')
var6_results = VAR(data3).fit(maxlags=6)
var6_results.summary()
```

8. Now set maxlags to 12 and use AIC and BIC to pick the optimum number of lags.

```python
# Use AIC
var12_results = VAR(data3).fit(maxlags=12)
aic = var12_results.aic
print('AIC : ' + str(aic))
```

```python
# Use BIC
bic = var12_results.bic
print('BIC : ' + str(bic))
```

9. For this question and the next one, use any of the above models. Test for Granger causality (you can use the test_causality() method):

(i) Test for $re \rightarrow rd$, that is, does equity return affect dividend return? 

(ii) Test for $rd \rightarrow re$, that is, does dividend return affect equity return? 

Write your conclusions.

```python
# Test causality (i)
# using 6 lags
granger_var6_results = VAR(data3).fit(maxlags=6)
# re -> rd
granger_causality_retord = granger_var6_results.test_causality(['Equity'], 'Dividends', kind='f')
print("Granger Causality Test: r_e -> r_d")
print(granger_causality_retord.summary())
```

```python
# Test causality (ii)
# rd -> re
granger_causality_rdtore = granger_var6_results.test_causality(['Dividends'], 'Equity', kind='f')
print("Granger Causality Test: r_d -> r_e")
print(granger_causality_rdtore.summary())
```

From both test, we reject the null hypothesis that equity return can affect dividend return and dividend return can affect equity return. The past values of equity returns do not contain useful information for predicting dividend returns, also the past values of dividend returns do not contain useful information for predicting equity returns.
