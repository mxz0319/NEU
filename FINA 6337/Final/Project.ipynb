```python
import numpy as np
import statsmodels.api as sm
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import norm, lognorm, binom
from statsmodels.distributions.empirical_distribution import ECDF
```

#### Brownian Motion

```python
def gen_bm(dt = 0.0001, T = 1):
    t = np.arange(0, 1+dt/2, dt)
    w = np.zeros(len(t))
    # every increament need to have random cosnt
    for i in range(1,len(w)):
        w[i] = w[i-1]+np.random.normal(scale = np.sqrt(dt))
    return w

def first_hitting(w, m):
    for i in range(len(w)):
        if w[i] >= m or w[i] <= -m:
            return i
    return 0
```

```python
M = [0.5,1] ; dt = 1/10000
for m in M:
    print('M = ' + str(m))
    hitting_time = []
    for _ in range(10000):
        w = gen_bm(dt = dt, T = 1)
        t_m = first_hitting(w, m)
        hitting_time.append(t_m)
    # ECDF
    ecdf = ECDF(hitting_time)
    #probs = ecdf(hitting_time)
    probs = 2 * norm.cdf(-1 * abs(m) / np.sqrt(ecdf.x[:10000]))
    # Plot
    plt.plot(hitting_time,probs)
    plt.show()
```

#### Binomial Model

```python
# stock price tree
def gen_stock_tree(s_0, sigma, T, n):
    dt = T/n
    u = np.exp(sigma*np.sqrt(dt)); d = 1/u
    S = np.zeros((n+1,n+1))
    S[0,0] = s_0
    for j in range(1, n+1):
        S[:,j] = S[:,j-1]*u
        S[j,j] = S[j-1,j-1]*d
    return (u, d, S)

def us_put(s_0, K, rf, sigma, T, n):
    dt = T/n
    t = np.linspace(0, T, n, endpoint = True)
    u,d,S = gen_stock_tree(s_0, sigma, T, n)
    q = (np.exp(rf*dt)-d)/(u-d)
    P = np.zeros((n+1,n+1))
    P[:,n] = np.clip(K-S[:,n], 0, None)
    # time & going backward & j is x
    for j in range(n-1, -1, -1):
        # i is y
        for i in range(j+1):
            P[i,j] = max(np.exp(-rf*dt)*(q*P[i,j+1]+(1-q)*P[i+1,j+1]), K-S[i,j])
            # d. set to 
            if j > 0 and P[i,j] < P[i,j - 1]:
                P[i,j] = P[i,j - 1]
    #return (t,P)
    return P[0,0]
```

```python
# Info.
s_0 = 100; K = 100; T = 1/4; rf = 0.10; sigma1 = 0.2; n1 = 100; sigma2 = 0.4; n2 = 1000;
```

```python
# sigma is 0.2
p1 = np.zeros(n1)
p2 = np.zeros(n2)
for n in range(n1):
    p1[n] = us_put(s_0, K, rf, sigma1, T, n+1)
for n in range(n2):
    p2[n] = us_put(s_0, K, rf, sigma1, T, n+1)
plt.plot(np.arange(1, n2+1), p2, label = 'n = 1000', color = 'orange')
plt.plot(np.arange(1, n1+1), p1, label = 'n = 100', color = 'red')
plt.legend()
```
